```python
count = 0
t = 0

# Get the headers
header = []

# Get the header row numbers
row_nheader = []
with open(folder + file) as f:
    for row in csv.reader(f, delimiter=',', skipinitialspace=True):
        count = count + 1
        if count <= number_of_lines_to_skip_in_the_beginning:
            continue
        
        # Check if the last entry of row can be a string
        # If the entry is floatable - that can read by pandas
        
        if len(row) > 1:
            try:
                last_entry = float(row[-1])
            except ValueError:
                header.append(row)
                row_nheader.append(count)
```

```python
def get_secs(x):
    hhss = x.split(":")
    seconds = 60*float(hhss[0]) + float(hhss[1])
    return seconds
```

## Read separate snapshots as a new dataframe

```python
df = []
for i in range(0, int(len(header)/3)):
    print(i)
    
    if i == int(len(header)/3) - 1:
        d = pd.read_csv(folder + file, skiprows = row_nheader[(i*3)+1]+1, 
                    header=None)
    else:
        d = pd.read_csv(folder + file, skiprows = row_nheader[(i*3)+1]+1, 
                        nrows= row_nheader[(i+1)*3] - row_nheader[(i*3)+1]-10 , header=None)

    d.columns = header[(3*i)+1]
    d['Sample Time'] = d['Sample Time'].apply(get_secs)
    df.append(d)
    
    datetime_object = datetime.strptime(header[i*3][1], '%m/%d/%Y %I:%M:%S %p',)
    
    d['Sample Time'] = d['Sample Time'] + datetime_object.timestamp()
    d.to_csv(folder + "TechStream_" + header[i*3][1].replace(' ', '_').replace('/','_')+ '.csv', index=False)
```
